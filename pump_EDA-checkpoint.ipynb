# Data cleaning 
There are many zero values in 'amount_tsh','gps_height','Population','longitude' and 'construction_year' columns.

```python
sum(df.amount_tsh==0)# most of the values(41639) in 'amount_tsh' is 0
df['amount_tsh'].replace(0.0, np.nan, inplace=True)#replaced all 0 values with NaN values.
df['amount_tsh'] = df.groupby('region_code').transform(lambda x: x.fillna(x.mean())).amount_tsh
#df['amount_tsh'].head(10)
```

```python
sum(df.gps_height==0)# Half of the values(20438) in 'gps_height' is 0
df['gps_height'].replace(0.0, np.nan, inplace=True)#replaced all 0 values with NaN values.
df['gps_height'] = df.groupby('region_code').transform(lambda x: x.fillna(x.mean())).gps_height
#df['gps_height'].head(10)
```

```python
sum(df.population==0)# most of the values(21381) in 'population' is 0
df['population'].replace(0.0, np.nan, inplace=True)#replaced all 0 values with NaN values.
df['population'] = df.groupby('region_code').transform(lambda x: x.fillna(x.mean())).population
#df['population'].head(10)
```

```python
sum(df.construction_year==0)# Half of the values(20709) in 'construction_year' is 0
df['construction_year'].replace(0.0, np.nan, inplace=True)#replaced all 0 values with NaN values.
df['construction_year'] = df.groupby('region_code').transform(lambda x: x.fillna(x.mean())).construction_year
#df['construction_year'].head(10)
```

```python
sum(df.longitude==0)# few values(1812) in 'longitude' is 0
df['longitude'].replace(0.0, np.nan, inplace=True)#replaced all 0 values with NaN values.
df['longitude'] = df.groupby('region_code').transform(lambda x: x.fillna(x.mean())).longitude
#df['longitude'].head(10)
```

# Handling Missing values in funder,scheme_management and wpt_name column.

```python
#df.funder.isnull().sum()# few valuse(3635) are missing.
#df.funder.value_counts()
def funder_m(row):  
    '''Keep top 5 values and set the rest to 'other'''

    if row['funder']=='Government Of Tanzania':
        return 'Gov'
    elif row['funder']=='Danida':
        return 'danida'
    elif row['funder']=='Hesawa':
        return 'hesawa'
    elif row['funder']=='Rwssp':
        return 'rwssp'
    elif row['funder']=='World Bank':
        return 'world_bank'    
    else:
        return 'other'
    
df['funder'] = df.apply(lambda row: funder_m(row), axis=1)
```

```python
def scheme_m(row):
    if row['scheme_management']=='VWC':
        return 'vwc'
    elif row['scheme_management']=='WUG':
        return 'wug'
    elif row['scheme_management']=='Water authority':
        return 'water authority'
    elif row['scheme_management']=='WUA':
        return 'wua'
    elif row['scheme_management']=='Water Board':
        return 'water Board'   
    elif row['scheme_management']=='parastatal':
        return 'parastatal'   
    elif row['scheme_management']=='Private operator':
        return 'private operator'  
    elif row['scheme_management']=='Company':
        return 'company'  
    else:
        return 'other'
df['scheme_management'] = df.apply(lambda row: scheme_m(row), axis=1)
```

```python
def extraction_type(row):
    if row['extraction_type']=='gravity':
        return 'Gravity'
    elif row['extraction_type']=='nira/tanira':
        return 'Nira/tanira '
    elif row['extraction_type']=='submersible':
        return 'Submersible'
    elif row['extraction_type']=='swn 80':
        return 'Swn 80'
    elif row['extraction_type']=='mono':
        return 'Mono'   
    elif row['extraction_type']=='india mark ii':
        return 'India mark ii'   
    elif row['extraction_type']=='afridev':
        return 'Afridev'  
    elif row['extraction_type']=='ksb':
        return 'ksb'  
    else:
        return 'Other'
df['extraction_type'] = df.apply(lambda row: extraction_type(row), axis=1)
```

```python
#df.wpt_name.value_counts()
#df.wpt_name.isnull().sum()
# Due to the huge number of factors and the lack of a clear dominating value drop 'wpt_name' column.
df = df.drop('wpt_name', axis=1)
```

```python
def ecdf(df):
    n=len(df)
    x=np.sort(df)
    y=np.arange(1,n+1)/n
    return x,y
```

```python
x,y=ecdf(df.amount_tsh)
plt.plot(x,y,marker='.',linestyle='none')
plt.xlabel('amount_tsh')
plt.ylabel('ECDF')
plt.show()
```
