# Introduction

Task:  Predict which water pumps are faulty? (predict which pumps are functional, which need some repairs, and which don't work at all?)
Using data from Taarifa and the Tanzanian Ministry of Water,  Predict one of these three classes based on a number of variables about what kind of pump is operating, when it was installed, and how it is managed. A smart understanding of which waterpoints will fail can improve maintenance operations and ensure that clean, potable water is available to communities across Tanzania.

Goal is to predict the operating condition of a waterpoint for each record in the dataset. 


```python
from IPython.display import Image
figsize=(30,8)
Image("picture.png",width=800,height=600)
```


Data: Using data from Taarifa and the Tanzanian Ministry of Water,  Data was collected using paper reports and feedback via phone calls. The data set has 40 features, like water quality, gps_hight, region code, extraction type and location etc.
The data set has 59400 entries and 41 features including the target column. The target column has three values (functional, functional-needs repair and non-functional). Most of the features are categorical.


Possible questions need to be answer.

1.Individual counts of Functional,Non functional and Functional Needs to repair.

2.Distribution of amount_tsh and gps_height among different status_grops.

3.Impact of population in different status_group.

4.Identify the region which contains most of the pumps.

5.Significance of categorical features(water_quality, extraction_type,waterpoint_type, payment, source,source_class,basin, funder, scheme_management,management,quantity) in the prediction of status_group.


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import sys
import os
```

```python
df1=pd.read_csv('Train.csv')
df2=pd.read_csv('Train_label.csv')
df=df1.merge(df2,on='id')
df.head()
```

```python
#all the values in num_private column is 0,deleting num_private column
df.drop(['num_private','scheme_name'],axis=1,inplace=True)
```

```python
#other columns have very similar data,so the corelation among those are very high.
#removing thouse columns to avoid risk of overfitting.
df.drop(['date_recorded','installer','subvillage','region','district_code','lga',
        'ward','public_meeting','permit','recorded_by','extraction_type_group','extraction_type_class',
        'management_group','payment_type','quality_group','quantity_group','source_type',
        'waterpoint_type_group'],axis=1,inplace=True)
```

# Data cleaning
There are many zero values in 'amount_tsh','gps_height','Population','longitude' and 'construction_year' columns.

```python
sum(df.amount_tsh==0)# most of the values(41639) in 'amount_tsh' is 0
df['amount_tsh'].replace(0.0, np.nan, inplace=True)#replaced all 0 values with NaN values.
df['amount_tsh'] = df.groupby('region_code').transform(lambda x: x.fillna(x.mean())).amount_tsh
df.dropna(subset=['amount_tsh'],axis=0,inplace=True)
#df['amount_tsh'].head(10)
```

```python
sum(df.gps_height==0)# Half of the values(20438) in 'gps_height' is 0
df['gps_height'].replace(0.0, np.nan, inplace=True)#replaced all 0 values with NaN values.
df['gps_height'] = df.groupby('region_code').transform(lambda x: x.fillna(x.mean())).gps_height
df.dropna(subset=['gps_height'],axis=0,inplace=True)
#df['gps_height'].head(10)
```

```python
sum(df.population==0)# most of the values(21381) in 'population' is 0
df['population'].replace(0.0, np.nan, inplace=True)#replaced all 0 values with NaN values.
df['population'] = df.groupby('region_code').transform(lambda x: x.fillna(x.mean())).population
df.dropna(subset=['population'],axis=0,inplace=True)
#df['population'].head(10)
```

```python
sum(df.construction_year==0)# Half of the values(20709) in 'construction_year' is 0
df['construction_year'].replace(0.0, np.nan, inplace=True)#replaced all 0 values with NaN values.
df['construction_year'] = df.groupby('region_code').transform(lambda x: x.fillna(x.mean())).construction_year
df.dropna(subset=['construction_year'],axis=0,inplace=True)
#df['construction_year'].head(10)
```

```python
sum(df.longitude==0)# few values(1812) in 'longitude' is 0
df['longitude'].replace(0.0, np.nan, inplace=True)#replaced all 0 values with NaN values.
df['longitude'] = df.groupby('region_code').transform(lambda x: x.fillna(x.mean())).longitude
df.dropna(subset=['longitude'],axis=0,inplace=True)
#df['longitude'].head(10)
```

# Handling Missing values in funder,scheme_management and wpt_name column.

```python
#df.funder.isnull().sum()# few valuse(3635) are missing.
#df.funder.value_counts()
def funder_m(row):  
    '''Keep top 5 values and set the rest to 'other'''

    if row['funder']=='Government Of Tanzania':
        return 'Gov'
    elif row['funder']=='Danida':
        return 'danida'
    elif row['funder']=='Hesawa':
        return 'hesawa'
    elif row['funder']=='Rwssp':
        return 'rwssp'
    elif row['funder']=='World Bank':
        return 'world_bank'    
    else:
        return 'other'
    
df['funder'] = df.apply(lambda row: funder_m(row), axis=1)
```

```python
def scheme_m(row):
    if row['scheme_management']=='VWC':
        return 'vwc'
    elif row['scheme_management']=='WUG':
        return 'wug'
    elif row['scheme_management']=='Water authority':
        return 'water authority'
    elif row['scheme_management']=='WUA':
        return 'wua'
    elif row['scheme_management']=='Water Board':
        return 'water Board'   
    elif row['scheme_management']=='parastatal':
        return 'parastatal'   
    elif row['scheme_management']=='Private operator':
        return 'private operator'  
    elif row['scheme_management']=='Company':
        return 'company'  
    else:
        return 'other'
df['scheme_management'] = df.apply(lambda row: scheme_m(row), axis=1)
```

```python
def extraction_type(row):
    if row['extraction_type']=='gravity':
        return 'Gravity'
    elif row['extraction_type']=='nira/tanira':
        return 'Nira/tanira '
    elif row['extraction_type']=='submersible':
        return 'Submersible'
    elif row['extraction_type']=='swn 80':
        return 'Swn 80'
    elif row['extraction_type']=='mono':
        return 'Mono'   
    elif row['extraction_type']=='india mark ii':
        return 'India mark ii'   
    elif row['extraction_type']=='afridev':
        return 'Afridev'  
    elif row['extraction_type']=='ksb':
        return 'ksb'  
    else:
        return 'Other'
df['extraction_type'] = df.apply(lambda row: extraction_type(row), axis=1)
```

```python
#df.wpt_name.value_counts()
#df.wpt_name.isnull().sum()
# Due to the huge number of factors and the lack of a clear dominating value drop 'wpt_name' column.
df = df.drop('wpt_name', axis=1)
```
